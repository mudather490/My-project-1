```python
import random

def choose_word():
    words = ["color", "banana", "football", "bad", "python", "machine"]
    return random.choice(words)


def show_hangman(lives):
    stages = [
r"""
   +---+
   |   |
   O   |
  /|\  |
  / \  |
       |
=========
""",
r"""
   +---+
   |   |
   O   |
  /|\  |
  /    |
       |
=========
""",
r"""
   +---+
   |   |
   O   |
  /|\  |
       |
       |
=========
""",
r"""
   +---+
   |   |
   O   |
  /|   |
       |
       |
=========
""",
r"""
   +---+
   |   |
   O   |
   |   |
       |
       |
=========
""",
r"""
   +---+
   |   |
   O   |
       |
       |
       |
=========
""",
r"""
   +---+
   |   |
       |
       |
       |
       |
=========
"""
]

    print(stages[lives])


def play_game():

    random_word = choose_word()

    guess = ["_" for letter in random_word]

    lives = 6
    used_letters = []

    print("\n🎮 HANGMAN GAME")
    print(guess)

    while "_" in guess and lives > 0:

        letter = input("\nPlease guess a letter: ").lower()

        if letter in used_letters:
            print("You already tried this letter!")
            continue

        used_letters.append(letter)

        if letter not in random_word:
            lives -= 1
            print("Wrong guess ❌")
            print("Lives left:", lives)
            show_hangman(lives)

        for place in range(len(random_word)):
            if random_word[place] == letter:
                guess[place] = letter

        print(guess)
        print("Used letters:", used_letters)


    if lives == 0:
        print("\n😢 You Lose!")
        print("The word was:", random_word)

    else:
        print("\n🎉 You Win!")
        score = lives * 10
        print("Your score:", score)


while True:

    play_game()

    again = input("\nPlay again? (y/n): ").lower()

    if again != "y":
        print("Thanks for playing 👋")
        break
```
